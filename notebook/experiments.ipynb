```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
google_api_key = os.getenv("GOOGLE_API_KEY")

if google_api_key == "":
    print("api key not found")
else:
    print(google_api_key)
```

```python
from llama_index.core import Settings, VectorStoreIndex
from QA_PDF.data_injection import load_custom_data
from QA_PDF.embedding import download_gemini_embedding
from QA_PDF.model_api import load_gemini_pro

# Bind components together inside global settings
Settings.llm = load_gemini_pro(google_api_key)
Settings.embed_model = download_gemini_embedding(google_api_key)

# Read documents, automatically chunk content, and index
docs = load_custom_data("./data")
index = VectorStoreIndex.from_documents(docs)

# Persist files locally onto your machine to save API resource consumption later
index.storage_context.persist(persist_dir="./storage")
```
